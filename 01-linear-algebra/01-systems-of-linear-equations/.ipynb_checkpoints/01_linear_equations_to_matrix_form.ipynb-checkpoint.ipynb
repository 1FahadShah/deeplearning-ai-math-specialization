# Systems of Linear Equations

In this exercise, we solve the following system of equations:

$$
\begin{cases}
5x + 3y + 5z = 6 \\
7x + 2y + 8z = 17 \\
4x + 0y + 3z = 8
\end{cases}
$$

We aim to represent this in **matrix form**:

$$
A \cdot X = B
$$

and solve for \(X\) using Python.


## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Define A and B

```python
# Coefficient matrix A

A = np.array([
    [5, 3, 5],
    [7, 2, 8],
    [4, 0, 3]
])

# Constants vector B
B = np.array([6, 17, 8])

print("Coefficient Matrix A:\n", A)
print("\nConstants Vector B:\n", B)
```

## Solve for X

```python
# Solve the system A·X = B
X = np.linalg.solve(A, B)
print("Solution Vector X:\n", X)
```

## Display Solution Clearly

```python
print(f"x = {X[0]:.2f}")
print(f"y = {X[1]:.2f}")
print(f"z = {X[2]:.2f}")
```

## Verify the Solution

```python
# Check if A·X ≈ B
is_correct = np.allclose(A @ X, B)
print("Verification Passed:", is_correct)
```

## Visualize 3D Plane Intersections

```python
# Create 3D plot for visualizing planes
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-2, 3, 10)
y = np.linspace(-2, 3, 10)
X_, Y_ = np.meshgrid(x, y)

# Plane equations: Ax + By + Cz = D
Z1 = (6 - 5*X_ - 3*Y_) / 5
Z2 = (17 - 7*X_ - 2*Y_) / 8
Z3 = (8 - 4*X_ - 0*Y_) / 3

ax.plot_surface(X_, Y_, Z1, alpha=0.5, rstride=100, cstride=100, color='red')
ax.plot_surface(X_, Y_, Z2, alpha=0.5, rstride=100, cstride=100, color='blue')
ax.plot_surface(X_, Y_, Z3, alpha=0.5, rstride=100, cstride=100, color='green')

# Intersection point
ax.scatter(X[0], X[1], X[2], color='black', s=50, label="Solution Point")

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title("Intersection of 3 Planes — Solution Point")
plt.show()
```
